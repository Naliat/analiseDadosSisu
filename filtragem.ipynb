```python
import pandas as pd

# Esse codigo foi criado com o intuito de filtrar os dados do dataset do sisu
df = pd.read_csv('lista_de_espera_sisu_2023_2.csv', sep='|', encoding='latin-1')
for col in ['NOTA_CANDIDATO', 'NOTA_CORTE']:
    df[col] = pd.to_numeric(df[col].astype(str).str.replace(',', '.'), errors='coerce')
df_filtrado = df[df['NOTA_CANDIDATO'] >= (df['NOTA_CORTE'] - 50)].copy()

# Definição de Alvo e a Variável de Distância
df_filtrado['EFETIVOU'] = df_filtrado['MATRICULA'].apply(lambda x: 1 if x == 'MATRÍCULA EFETIVADA' else 0)
df_filtrado['MESMO_ESTADO'] = (df_filtrado['UF_CANDIDATO'] == df_filtrado['UF_IES']).astype(int)

# Define a quantidade de variavel, nesse caso foi 1500
df_final = df_filtrado.sort_values(by='NOTA_CANDIDATO', ascending=False).head(1500)

df_final.to_csv('sisu_preparado.csv', index=False)
print(f"✅ Filtro concluído! {len(df_final)} candidatos competitivos salvos para o Julia.")
```
